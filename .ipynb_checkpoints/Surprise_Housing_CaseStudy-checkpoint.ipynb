### Outlier check and visualization

```python
# function to remove outlier 
def findremout(dataseries):
    q1=dataseries.quantile(.25)
    q3=dataseries.quantile(.75)
    iqr=q3-q1
    upoutlierind=np.where(dataseries>=(q3+1.5*iqr))
    lwoutlierind=np.where(dataseries<=(q1-1.5*iqr))
    return [upoutlierind,lwoutlierind]
```

```python
def dataOverview(dataframe,dataSeries,bin=5):
    print("Data for :"+dataSeries)
    print("Describe Data: \n",dataframe[dataSeries].describe().round(2))
    print("ValueCount Data: \n",dataframe[dataSeries].value_counts(bins=bin).round(2))
    plt.figure(figsize=(15,10))
    plt.subplot(221)
    sns.boxplot(dataframe[dataSeries])
    plt.title(dataSeries)
    plt.subplot(222)
    sns.distplot(dataframe[dataSeries])
    plt.title(dataSeries)
    plt.show() 
# this function will generate data related to specific attribute and helps in analyzing the data by drawing boxplot and histogram    
```

```python
# checking outliers in all num columns
for col in df_house.select_dtypes(exclude='object').columns:
    dataOverview(df_house,col)
```

##### Splitting the Data

```python
y=df_house.pop('SalePrice')
X=df_house
```

```python
print(y.head())
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,train_size=0.7,random_state=100)
```

```python
# Numeric columns to be scaled 
# Scaling using MinMax Scaler
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
cols=X_train.select_dtypes(include=['int32','int64','float64','float32']).columns
X_train[cols]=scaler.fit_transform(X_train[cols])
X_test[cols]=scaler.transform(X_test[cols])
```

```python
X_train.head()
```

```python
X_train.describe()
```

##### Model Building and Evaluation using RFE

```python
from sklearn.feature_selection import RFE
```

```python
#linear regression
lm=LinearRegression()
lm.fit(X_train,y_train)
# RFE
rfe=RFE(lm,25)
rfe=rfe.fit(X_train,y_train)
```

```python
#top features identified by RFE
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
column=X_train.columns[rfe.support_]
```

```python
column
```

```python
X_train1=X_train[column]
X_test_rfe1=X_test[column]
```

```python
lm1=lm.fit(X_train,y_train)
#Coefficent and intercept
print(lm1.intercept_)
print(lm1.coef_)
```

```python
# finding r2_score , RSS and RMSE
from sklearn.metrics import r2_score, mean_squared_error
```

```python
y_prd_train=rfe.predict(X_train)
y_pred_test=rfe.predict(X_test)
```

```python
r2_trn =r2_score(y_train,y_prd_train)
print("R2 -Train :",r2_trn)
r2_tst=r2_score(y_test,y_pred_test)
print("R2- Test :",r2_tst)
rss_trn=np.sum(np.square(y_train-y_prd_train))
rss_tst=np.sum(np.square(y_test-y_pred_test))
print("RSS-Train",rss_trn)
print("RSS-Test",rss_tst)
mse_trn=mean_squared_error(y_train,y_prd_train)
mse_test=mean_squared_error(y_test,y_pred_test)
print("MSE-Train",mse_trn)
print("MSE-Test",mse_test)
```
